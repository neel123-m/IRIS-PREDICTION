```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Irish csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
x
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder() 
df['species']    #access from dataset different coloumn
df['new']=le.fit_transform(df['species']) #adding a new coloumn to numerical value
df
```

```python
x=df.drop(['species','new'],axis=1) #drop 2 coloumns 
x
```

```python
y=df['new']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
xtrain


```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(xtrain,ytrain)
RF.score(xtest,ytest)
```

```python
from sklearn.linear_model import LogisticRegression
LF=LogisticRegression()
LF.fit(xtrain,ytrain)
LF.score(xtest,ytest)
```

```python
from sklearn import svm
SV=svm.SVC()
SV.fit(xtrain,ytrain)
SV.score(xtest,ytest)
```

```python
#k fold cross validation
from sklearn.model_selection import cross_val_score
cv=cross_val_score(RF,x,y,cv=15)
cv
cv.mean()
```

```python
from sklearn.model_selection import cross_val_score
cv=cross_val_score(LF,x,y,cv=10)
cv
cv.mean()
```

```python
#k fold cross validation
from sklearn.model_selection import cross_val_score
cv=cross_val_score(SV,x,y,cv=10)
cv
cv.mean()
```
